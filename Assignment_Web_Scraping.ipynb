```python
#Q10. 10) Write a python program to scrape the details of top publications from Google Scholar from https://scholar.google.com/citations?view_op=top_venues&hl=en
# i) Rank, ii) Publication, iii) h5-index,  iv) h5-median

page10=requests.get("https://scholar.google.com/citations?view_op=top_venues&hl=en")
page10
```

```python
rank=[]
for i in soup10.find_all('td',  class_="gsc_mvt_p"):
    rank.append(i.text)
rank
```

```python
publication=[]
for i in soup10.find_all('td', class_="gsc_mvt_t"):
    publication.append(i.text)
publication
```

```python
h5_ind=[]
temp=[]
for i in soup10.find_all('td', class_="gsc_mvt_n"):
    temp.append(i.text)
for i in range(0,len(temp),2):
    h5_ind.append(temp[i])
h5_ind
```

```python
h5_med=[]
for i in range(1,len(temp),2):
    h5_med.append(temp[i])
h5_med
```

```python
print(len(rank), len(publication), len(h5_ind), len(h5_med))
```

```python
df10= pd.DataFrame({'Rank':rank, 'Publications': publication, 'H5-Index': h5_ind, 'h5-Median':h5_med})
print("Details of Top Publications From Google Scholars")
df10
```
